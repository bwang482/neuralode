## Age distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_cas = df_train[df_train['label']==1]
df_ctl = df_train[df_train['label']==0]

fig, ax = plt.subplots(figsize=(8, 6))

sns.kdeplot(df_cas['age'], color='skyblue', label='MDD cases', 
            linewidth=2, fill=True, alpha=0.3, ax=ax)
sns.kdeplot(df_ctl['age'], color='salmon', label='MDD controls', 
            linewidth=2, fill=True, alpha=0.3, ax=ax)

ax.legend()
ax.set_xlabel('Age')
ax.set_ylabel('Density')
plt.show()
```

## Length of (sampled) EHR history

```python
fig, ax = plt.subplots(figsize=(8, 6))

num_days_cas = df_cas['all_dates'].apply(
    lambda x: (x[-1] - x[0]).days if x else None
).dropna().tolist()
num_days_ctl = df_ctl['all_dates'].apply(
    lambda x: (x[-1] - x[0]).days if x else None
).dropna().tolist()

bins = np.histogram_bin_edges(np.concatenate([num_days_cas, num_days_ctl]), bins=20)

sns.histplot(num_days_cas, bins=bins, color='#4A90E2', label='MDD cases',
            alpha=0.6, edgecolor='black', linewidth=0.5,
            stat='count', ax=ax)
sns.histplot(num_days_ctl, bins=bins, color='#F5715C', label='MDD controls',
            alpha=0.6, edgecolor='black', linewidth=0.5,
            stat='count', ax=ax)

ax.legend()
ax.set_xlabel('Length of EHR history (in days)')
ax.set_ylabel('Density')
plt.show()
```

## Number of (sampled) visits

```python
fig, ax = plt.subplots(figsize=(8, 6))

num_tps_cas = df_cas['all_dates'].apply(
    lambda x: len(x) if x else None
).dropna().tolist()
num_tps_ctl = df_ctl['all_dates'].apply(
    lambda x: len(x) if x else None
).dropna().tolist()

bins = np.histogram_bin_edges(np.concatenate([num_tps_cas, num_tps_ctl]), bins=20)

sns.histplot(num_tps_cas, bins=bins, color='#4A90E2', label='MDD cases',
            alpha=0.6, edgecolor='black', linewidth=0.5,
            stat='count', ax=ax)
sns.histplot(num_tps_ctl, bins=bins, color='#F5715C', label='MDD controls',
            alpha=0.6, edgecolor='black', linewidth=0.5,
            stat='count', ax=ax)

ax.legend()
ax.set_xlabel('Number of time points (visits)')
ax.set_ylabel('Density')
plt.show()
```

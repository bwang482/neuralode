```python
import os
import argparse
import numpy as np
from loguru import logger
from datetime import datetime

import torch
from torch.distributions.normal import Normal

from config import random_state, MODELS_DIR, RESULTS_DIR
from utils.general_functs import save_pickle, load_pickle
from utils.viz_functs import parse_model_filename, format_model_filename
import utils.model_functs as utils
```

```python
parser = argparse.ArgumentParser(description = 'Latent ODE Inference')

# Dataset and task parameters
parser.add_argument('--dataset', type = str, default = 'MDD', help = "dataset to load, available: MDD, CAD")
parser.add_argument('--look_back', type = str, default = '3y', help = "look back window (1y, 2y, 3y)")
parser.add_argument('--feat_type', type = str, default = 'demo_dx_med', help = "Feature type (demo_dx, demo_dx_med)")
parser.add_argument('--sample_tp', type = float, default = None, help = "number/percentage of time points to sub-sample")
parser.add_argument('--extrap', action = 'store_true', help = "use extrapolation mode (default: True)")
parser.add_argument('--batch_size', type = int, default = 128, help ="batch size")
parser.add_argument('--n_iter', type = int, default = 10, help = "number of iterations")
parser.add_argument('--warmup_epochs', type=int, default=0, help="Number of epochs for classifier warm-up")

# Training and model parameters
parser.add_argument('--gen-latent-dims', type = int, default = 100, help = "dimensionality of the latent state in the generative ODE")
parser.add_argument('--rec-latent-dims', type = int, default = 100, help = "dimensionality of the latent state in the recognition ODE")
parser.add_argument('--gen-layers', type = int, default = 3, help = "number of layers in ODE func in generative ODE")
parser.add_argument('--rec-layers', type = int, default = 3, help = "number of layers in ODE func in recognition ODE")
parser.add_argument('--units', type = int, default = 100, help = "number of units per layer in ODE func")
parser.add_argument('--gru-units', type = int, default = 100, help = "number of units per layer in the GRU update network")
parser.add_argument('--classif-units', type = int, default = 100, help = "number of units per layer in the classification network")
parser.add_argument('--classif', action = 'store_false', help = "include binary classification loss")
parser.add_argument('--classif_w_recon', action = 'store_true', help = "jointly consider classification loss and reconstruction loss")
# parser.add_argument('--dropout', type = float, default = 1.0, help = "Dropout rate for classifier (default: 1.0)")
# parser.add_argument('--classifier-layers', type = int, default = 2, help = "Number of hidden layers in classifier (default: 3)")
parser.add_argument('--feat_dir', type = str, default = 'random_visit_w_filters_all_matched', help = "")

# Infra parameters
parser.add_argument('--cuda', type=str, default='cuda:0', help="which gpu to use")

args, unkwn = parser.parse_known_args()
args.random_seed = random_state

device = torch.device(args.cuda if torch.cuda.is_available() else "cpu")
torch.set_default_device(args.cuda)
torch.set_default_dtype(torch.float32)

# Set seed for reproducibility
torch.manual_seed(args.random_seed)
np.random.seed(args.random_seed)
```

## Visualization

```python
def visualize_predictions(model, data_batch, args, n_traj=1, save_dir=None, display_in_notebook=True):
    """
    Visualize model predictions for time series data with support for Jupyter notebook display
    
    Args:
        model: The trained LatentODE model
        data_batch: A batch of data from the dataloader
        args: Command line arguments
        n_traj: Number of trajectories to sample for reconstruction
        save_dir: Directory to save visualizations (only used if display_in_notebook=False)
        display_in_notebook: If True, displays plots in the notebook instead of saving files
    """
    import os
    import numpy as np
    import matplotlib.pyplot as plt
    
    if not display_in_notebook and save_dir is not None:
        os.makedirs(save_dir, exist_ok=True)
    
    if display_in_notebook:
        from IPython.display import display
        %matplotlib inline
    
    model.eval()
    with torch.no_grad():
        # Call get_recons to get predictions directly
        pred_y, info = model.get_recons(
            tps_to_pred=data_batch["tps_to_pred"],
            obs_data=data_batch["obs_data"],
            obs_tps=data_batch["obs_tps"],
            mask=data_batch["obs_mask"],
            n_traj=n_traj,
            z_last=args.z_last if hasattr(args, 'z_last') else False
        )
        
        # Extract data
        obs_tps_np = data_batch["obs_tps"].cpu().numpy()
        pred_tps_np = data_batch["tps_to_pred"].cpu().numpy()
        obs_data_np = data_batch["obs_data"].cpu().numpy()
        true_vals_np = data_batch["data_to_pred"].cpu().numpy()
        
        # Convert predictions to numpy
        if n_traj > 1:
            # Average predictions across trajectories
            pred_vals_np = torch.mean(pred_y, dim=0).cpu().numpy()
            # Keep all trajectories for confidence intervals
            all_trajs_np = pred_y.cpu().numpy()
        else:
            pred_vals_np = pred_y[0].cpu().numpy()
            all_trajs_np = None
        
        # Extract latent states
        fp_mu, fp_std, fp_enc = info["fp"]
        z0 = fp_enc.cpu().numpy()  # Initial latent states
        
        # Get latent trajectories if available
        latent_traj = None
        if "latent_traj" in info:
            latent_traj = info["latent_traj"].cpu().numpy()
        
        # Get labels and predictions
        has_labels = "labels" in data_batch and data_batch["labels"] is not None
        if has_labels:
            labels = data_batch["labels"].cpu().numpy()
        
        has_predictions = model.classif and "label_pred" in info
        if has_predictions:
            pred_probs = torch.sigmoid(info["label_pred"]).cpu().numpy()
        
        # Plot time series predictions
        n_samples = min(5, pred_vals_np.shape[0])  # Plot up to 5 samples
        n_features = min(5, pred_vals_np.shape[2])  # Plot up to 5 features
        
        for i in range(n_samples):
            plt.figure(figsize=(15, 3 * n_features))
            
            for j in range(n_features):
                plt.subplot(n_features, 1, j+1)
                
                # Plot ground truth for prediction times
                plt.plot(pred_tps_np, true_vals_np[i, :, j], 'b-', 
                         label='True values', linewidth=2, alpha=0.7)
                
                # Plot model predictions
                plt.plot(pred_tps_np, pred_vals_np[i, :, j], 'r--', 
                         label='Predicted', linewidth=2)
                
                # Plot confidence intervals if multiple trajectories
                if all_trajs_np is not None and n_traj > 1:
                    traj_values = all_trajs_np[:, i, :, j]
                    traj_mean = np.mean(traj_values, axis=0)
                    traj_std = np.std(traj_values, axis=0)
                    plt.fill_between(
                        pred_tps_np,
                        traj_mean - 2*traj_std,
                        traj_mean + 2*traj_std,
                        color='r', alpha=0.2, label='95% CI'
                    )
                
                # Add observed data points
                plt.scatter(obs_tps_np, obs_data_np[i, :, j], 
                            c='g', marker='o', label='Observed', s=30)
                
                # Set axis labels and legend
                plt.xlabel('Time')
                plt.ylabel(f'Feature {j+1}')
                plt.legend()
                
                if j == 0:
                    # Add label information if available
                    title = f'Sample {i+1}'
                    if has_labels:
                        label_val = int(labels[i].item()) if hasattr(labels[i], 'item') else int(labels[i][0])
                        title += f' - True Class: {label_val}'
                    if has_predictions:
                        pred_val = pred_probs[0, i].item() if n_traj > 0 else pred_probs[i].item()
                        title += f' - Pred Prob: {pred_val:.3f}'
                    plt.title(title)
            
            plt.tight_layout()
            
            if display_in_notebook:
                display(plt.gcf())  # Display in notebook
                plt.close()
            else:
                plt.savefig(os.path.join(save_dir, f'sample_{i+1}_prediction.png'))
                plt.close()
        
        # Plot latent space visualization if at least 2D
        if z0.shape[-1] >= 2:
            plt.figure(figsize=(10, 8))
            
            # Use the first trajectory for visualization
            z0_vis = z0[0] if z0.ndim > 2 else z0
            
            if has_labels:
                # Color by class if labels are available
                unique_labels = np.unique(labels)
                for label in unique_labels:
                    if labels.ndim > 1:
                        mask = labels[:, 0] == label
                    else:
                        mask = labels == label
                    plt.scatter(z0_vis[mask, 0], z0_vis[mask, 1], 
                              label=f'Class {int(label)}', alpha=0.7)
            else:
                # Otherwise just show all points
                plt.scatter(z0_vis[:, 0], z0_vis[:, 1], alpha=0.7)
            
            plt.title('Latent Space Visualization (First 2 Dimensions)')
            plt.xlabel('Z0 Dimension 1')
            plt.ylabel('Z0 Dimension 2')
            plt.legend()
            
            if display_in_notebook:
                display(plt.gcf())
                plt.close()
            else:
                plt.savefig(os.path.join(save_dir, 'latent_space.png'))
                plt.close()
            
            # Plot latent trajectories if available
            if latent_traj is not None and latent_traj.shape[-1] >= 2:
                plt.figure(figsize=(10, 8))
                
                # Extract first 3 samples for clarity
                samples_to_plot = min(3, latent_traj.shape[1])
                for i in range(samples_to_plot):
                    # Get latent trajectory for this sample
                    sample_traj = latent_traj[0, i, :, :]  # First trajectory, sample i
                    
                    # Plot trajectory in 2D space
                    plt.plot(sample_traj[:, 0], sample_traj[:, 1], '-o', 
                             label=f'Sample {i+1}', markersize=4)
                    
                    # Mark start and end points
                    plt.scatter(sample_traj[0, 0], sample_traj[0, 1], 
                                marker='*', s=100, label=f'Start {i+1}' if i == 0 else "")
                    plt.scatter(sample_traj[-1, 0], sample_traj[-1, 1], 
                                marker='s', s=100, label=f'End {i+1}' if i == 0 else "")
                
                plt.title('Latent Trajectories in 2D Space')
                plt.xlabel('Latent Dimension 1')
                plt.ylabel('Latent Dimension 2')
                plt.legend()
                
                if display_in_notebook:
                    display(plt.gcf())
                    plt.close()
                else:
                    plt.savefig(os.path.join(save_dir, 'latent_trajectories.png'))
                    plt.close()
    
    if not display_in_notebook:
        print(f"Saved visualizations to {save_dir}")
    else:
        print("Visualizations displayed in notebook")
```

```python
test_batch = next(iter(test_dataloader))

# Run visualization
visualize_predictions(
    model,
    test_batch,
    args,
    n_traj=1,  
    display_in_notebook=True
)
```

```python
def visualize_latent_trajectories(model, data_batch, args, n_traj=1, n_samples=None, 
                                 use_pca=False, annotate_time=False, display_in_notebook=True, 
                                 save_path=None):
    """
    Visualize latent trajectories in 2D space, comparing patients with different outcomes.
    
    Args:
        model: The trained LatentODE model
        data_batch: A batch of data from the dataloader
        args: Command line arguments
        n_traj: Number of trajectories to sample 
        n_samples: Number of samples to visualize (default: all in batch)
        use_pca: Whether to use PCA for dimensionality reduction (if latent dim > 2)
        annotate_time: Whether to annotate time points on trajectories
        display_in_notebook: Whether to display in notebook
        save_path: Path to save figure
    """
    import numpy as np
    import matplotlib.pyplot as plt
    from matplotlib.lines import Line2D
    from matplotlib.patches import Patch
    import torch
    
    # For PCA if needed
    if use_pca:
        from sklearn.decomposition import PCA
    
    # Set plotting style
    plt.style.use('seaborn-v0_8-whitegrid')
    
    model.eval()
    with torch.no_grad():
        # Get latent trajectories from the model
        pred_y, info = model.get_recons(
            tps_to_pred=data_batch["tps_to_pred"],
            obs_data=data_batch["obs_data"],
            obs_tps=data_batch["obs_tps"],
            mask=data_batch["obs_mask"],
            n_traj=n_traj,
            z_last=args.z_last if hasattr(args, 'z_last') else False
        )
        
        # Extract latent trajectories and labels
        latent_traj = info["latent_traj"].cpu().numpy()
        
        # Get predictions and true labels
        has_labels = "labels" in data_batch and data_batch["labels"] is not None
        if has_labels:
            labels = data_batch["labels"].cpu().numpy()
            if labels.ndim > 1:
                labels = labels.squeeze()
        else:
            # If no labels, use model predictions
            if "label_pred" in info and model.classif:
                pred_probs = torch.sigmoid(info["label_pred"]).cpu().numpy().squeeze()
                labels = (pred_probs > 0.5).astype(int)
            else:
                labels = np.zeros(latent_traj.shape[1])
        
        # Limit number of samples if specified
        if n_samples is not None and n_samples < latent_traj.shape[1]:
            # Ensure we have balanced classes in the visualization
            pos_indices = np.where(labels == 1)[0]
            neg_indices = np.where(labels == 0)[0]
            
            n_pos = min(len(pos_indices), n_samples // 2)
            n_neg = min(len(neg_indices), n_samples - n_pos)
            
            selected_pos = np.random.choice(pos_indices, n_pos, replace=False)
            selected_neg = np.random.choice(neg_indices, n_neg, replace=False)
            
            selected_indices = np.concatenate([selected_pos, selected_neg])
            np.random.shuffle(selected_indices)
            
            latent_traj = latent_traj[:, selected_indices]
            labels = labels[selected_indices]
        
        # Get time points
        pred_tps_np = data_batch["tps_to_pred"].cpu().numpy()
        
        # Prepare for dimensionality reduction if needed
        latent_dim = latent_traj.shape[-1]
        
        if latent_dim > 2:
            if use_pca:
                # Reshape for PCA
                traj_shape = latent_traj.shape
                reshaped_traj = latent_traj.reshape(-1, latent_dim)
                
                # Apply PCA
                pca = PCA(n_components=2)
                reduced_traj = pca.fit_transform(reshaped_traj)
                
                # Reshape back
                reduced_traj = reduced_traj.reshape(traj_shape[0], traj_shape[1], traj_shape[2], 2)
                plotting_traj = reduced_traj
                
                # Get variance explained
                var_explained = pca.explained_variance_ratio_
                dim_labels = [f"PC1 ({var_explained[0]:.0%})", f"PC2 ({var_explained[1]:.0%})"]
            else:
                # Just use first two dimensions
                plotting_traj = latent_traj[:, :, :, :2]
                dim_labels = ["Latent Dimension 1", "Latent Dimension 2"]
        else:
            plotting_traj = latent_traj
            dim_labels = ["Latent Dimension 1", "Latent Dimension 2"]
        
        # Create the figure
        plt.figure(figsize=(6, 6))
        
        # Set up colors and markers
        colors = {0: 'blue', 1: 'red'}  # 0: No CAD, 1: CAD
        color_names = {0: 'Non-CAD', 1: 'CAD'}
        markers = {0: 'o', 1: 's'}
        
        # Set up the legend elements
        legend_elements = []
        for label_val in np.unique(labels):
            legend_elements.append(
                Line2D([0], [0], color=colors[label_val], lw=2, 
                       label=f'{color_names[label_val]} Trajectory')
            )
            legend_elements.append(
                Line2D([0], [0], marker='*', color='white', 
                       markerfacecolor=colors[label_val], markersize=15, 
                       label=f'{color_names[label_val]} Start')
            )
            legend_elements.append(
                Line2D([0], [0], marker=markers[label_val], color='white', 
                       markerfacecolor=colors[label_val], markersize=10, 
                       label=f'{color_names[label_val]} End')
            )
            
        # Plot each trajectory
        n_samples = plotting_traj.shape[1]
        for i in range(n_samples):
            label = int(labels[i])
            color = colors[label]
            
            # Extract trajectory for this sample (first trajectory)
            sample_traj = plotting_traj[0, i, :, :]
            
            # Plot the trajectory
            plt.plot(sample_traj[:, 0], sample_traj[:, 1], '-', 
                     color=color, alpha=0.7, linewidth=1.5)
            
            # Mark start point with star
            plt.scatter(sample_traj[0, 0], sample_traj[0, 1], 
                        marker='*', s=150, color=color, edgecolor='black', linewidth=0.5)
            
            # Mark end point with square or circle based on class
            plt.scatter(sample_traj[-1, 0], sample_traj[-1, 1], 
                        marker=markers[label], s=100, color=color, edgecolor='black', linewidth=0.5)
            
            # Annotate time points if requested
            if annotate_time and i < 3:  # Limit to first 3 samples to avoid clutter
                n_time_annotations = min(5, len(pred_tps_np))
                step = len(pred_tps_np) // n_time_annotations
                for t_idx in range(0, len(pred_tps_np), step):
                    if t_idx > 0 and t_idx < len(pred_tps_np) - 1:  # Skip first and last
                        plt.annotate(f"{pred_tps_np[t_idx]:.2f}", 
                                    (sample_traj[t_idx, 0], sample_traj[t_idx, 1]),
                                    fontsize=8, alpha=0.7)
        
        # Add density estimation to show class regions (optional)
        try:
            from scipy.stats import gaussian_kde
            
            # Extract end points
            end_points_x = plotting_traj[0, :, -1, 0]
            end_points_y = plotting_traj[0, :, -1, 1]
            
            # Create a mesh grid for the density plot
            x_min, x_max = plt.xlim()
            y_min, y_max = plt.ylim()
            margin = 0.1
            x_min -= margin * (x_max - x_min)
            x_max += margin * (x_max - x_min)
            y_min -= margin * (y_max - y_min)
            y_max += margin * (y_max - y_min)
            
            xx, yy = np.mgrid[x_min:x_max:100j, y_min:y_max:100j]
            positions = np.vstack([xx.ravel(), yy.ravel()])
            
            # Calculate density for each class
            for label_val in np.unique(labels):
                class_indices = labels == label_val
                if sum(class_indices) > 5:  # Need enough points for density estimation
                    class_points = np.vstack([
                        end_points_x[class_indices], 
                        end_points_y[class_indices]
                    ])
                    
                    try:
                        kernel = gaussian_kde(class_points)
                        z = np.reshape(kernel(positions).T, xx.shape)
                        
                        # Plot density contour
                        plt.contour(xx, yy, z, levels=3, colors=colors[label_val], 
                                   alpha=0.3, linestyles='dashed')
                    except:
                        # Skip if kde fails (can happen with too few points)
                        pass
        except:
            # Skip if scipy not available or any other error
            pass
        
        # Set labels and title
        plt.xlabel(dim_labels[0], fontsize=14)
        plt.ylabel(dim_labels[1], fontsize=14)
        plt.title("Latent Trajectories by Patient Outcome", fontsize=16)
        
        # Add legend
        plt.legend(handles=legend_elements, loc='upper right', fontsize=10)
        
        # Add a text box with summary statistics
        n_cad = sum(labels == 1)
        n_non_cad = sum(labels == 0)
        # plt.figtext(0.02, 0.02, 
        #            f"Total patients: {len(labels)}\nCAD patients: {n_cad} ({n_cad/len(labels):.1%})\n" +
        #            f"Non-CAD patients: {n_non_cad} ({n_non_cad/len(labels):.1%})",
        #            fontsize=10, bbox=dict(facecolor='white', alpha=0.7))
        
        plt.tight_layout()
        
        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
            print(f"Figure saved to {save_path}")
            
        if display_in_notebook:
            from IPython.display import display
            display(plt.gcf())
        
        plt.close()
```

```python
def visualize_latent_space_umap(model, dataloader, args, n_traj=1, 
                                trajectory_points='endpoints', n_neighbors=15, min_dist=0.1,
                                display_in_notebook=True, save_path=None):
    """
    Visualize latent states using UMAP dimensionality reduction.
    
    Args:
        model: The trained LatentODE model
        dataloader: DataLoader with patient data
        args: Command line arguments
        n_traj: Number of trajectories to sample
        trajectory_points: Which points to use from trajectories:
                          'endpoints' - just start and end points
                          'initial' - only initial states
                          'final' - only final states
        n_neighbors: UMAP parameter - controls how UMAP balances local versus global structure
        min_dist: UMAP parameter - controls how tightly UMAP packs points together
        display_in_notebook: Whether to display in notebook
        save_path: Path to save figure
    """
    import numpy as np
    import matplotlib.pyplot as plt
    from matplotlib.lines import Line2D
    import torch
    from tqdm import tqdm
    
    # Import UMAP
    try:
        import umap
    except ImportError:
        print("UMAP not installed. Please install with: pip install umap-learn")
        return
    
    # Set plotting style
    plt.style.use('seaborn-v0_8-whitegrid')
    
    model.eval()
    
    # Collect embeddings and labels
    all_initial_states = []
    all_final_states = []
    all_labels = []
    all_patient_ids = []
    
    print("Collecting latent representations from all patients...")
    batch_count = 0
    max_batches = 10  # Limit the number of batches for memory reasons
    
    with torch.no_grad():
        for batch_idx, data_batch in enumerate(tqdm(dataloader)):
            if batch_count >= max_batches:
                break
                
            # Call get_recons to get predictions and latent trajectories
            pred_y, info = model.get_recons(
                tps_to_pred=data_batch["tps_to_pred"],
                obs_data=data_batch["obs_data"],
                obs_tps=data_batch["obs_tps"],
                mask=data_batch["obs_mask"],
                n_traj=n_traj,
                z_last=args.z_last if hasattr(args, 'z_last') else False
            )
            
            # Extract initial latent states (z0) and latent trajectories
            fp_mu, fp_std, fp_enc = info["fp"]
            latent_traj = info["latent_traj"]
            
            # Extract initial and final states
            initial_states = fp_enc[0].cpu().numpy()  # (batch_size, latent_dim)
            final_states = latent_traj[0, :, -1, :].cpu().numpy()  # (batch_size, latent_dim)
            
            # Get labels
            if "labels" in data_batch and data_batch["labels"] is not None:
                labels = data_batch["labels"].cpu().numpy()
                if labels.ndim > 1:
                    labels = labels.squeeze()
            else:
                # If no labels, use model predictions
                if "label_pred" in info and model.classif:
                    pred_probs = torch.sigmoid(info["label_pred"]).cpu().numpy().squeeze()
                    labels = (pred_probs > 0.5).astype(int)
                else:
                    labels = np.zeros(fp_enc.shape[1])
            
            # Extract patient IDs if available
            if "patient_ids" in data_batch:
                patient_ids = data_batch["patient_ids"]
            else:
                # Create dummy IDs if not available
                patient_ids = [f"P{batch_idx}_{i}" for i in range(fp_enc.shape[1])]
            
            # Store data
            all_initial_states.append(initial_states)
            all_final_states.append(final_states)
            all_labels.append(labels)
            all_patient_ids.extend(patient_ids)
            
            batch_count += 1
    
    # Combine data from all batches
    all_initial_states = np.concatenate(all_initial_states, axis=0)
    all_final_states = np.concatenate(all_final_states, axis=0)
    all_labels = np.concatenate(all_labels, axis=0)
    
    # Prepare data for UMAP based on which points we want to use
    if trajectory_points == 'endpoints':
        # Use both initial and final states
        umap_data = np.vstack([all_initial_states, all_final_states])
        umap_labels = np.concatenate([all_labels, all_labels])
        point_types = np.array(['initial'] * len(all_labels) + ['final'] * len(all_labels))
    elif trajectory_points == 'initial':
        # Use only initial states
        umap_data = all_initial_states
        umap_labels = all_labels
        point_types = np.array(['initial'] * len(all_labels))
    elif trajectory_points == 'final':
        # Use only final states
        umap_data = all_final_states
        umap_labels = all_labels
        point_types = np.array(['final'] * len(all_labels))
    else:
        raise ValueError("trajectory_points must be 'endpoints', 'initial', or 'final'")
    
    print(f"Applying UMAP to {umap_data.shape[0]} data points with dimension {umap_data.shape[1]}...")
    
    # Apply UMAP for dimensionality reduction
    reducer = umap.UMAP(n_neighbors=n_neighbors, min_dist=min_dist, n_components=2, random_state=42)
    embedding = reducer.fit_transform(umap_data)
    
    # Create the visualization
    plt.figure(figsize=(14, 10))
    
    # Set up colors and markers
    colors = {0: 'blue', 1: 'red'}  # 0: No CAD, 1: CAD
    color_names = {0: 'Non-CAD', 1: 'CAD'}
    markers = {'initial': 'o', 'final': 's'}
    
    # Plot each point
    for label_val in np.unique(umap_labels):
        for point_type in np.unique(point_types):
            mask = (umap_labels == label_val) & (point_types == point_type)
            plt.scatter(embedding[mask, 0], embedding[mask, 1], 
                      color=colors[label_val], 
                      marker=markers[point_type],
                      alpha=0.7, 
                      label=f"{color_names[label_val]} ({point_type})",
                      s=50)
    
    # If we're using endpoints, draw lines between initial and final states
    if trajectory_points == 'endpoints':
        n_patients = len(all_labels)
        for i in range(n_patients):
            # Get indexes for this patient's initial and final state
            initial_idx = i
            final_idx = i + n_patients
            
            # Draw a line connecting them
            plt.plot([embedding[initial_idx, 0], embedding[final_idx, 0]], 
                    [embedding[initial_idx, 1], embedding[final_idx, 1]],
                    color=colors[all_labels[i]], alpha=0.3)
    
    # Set labels and title
    plt.xlabel('UMAP Dimension 1', fontsize=14)
    plt.ylabel('UMAP Dimension 2', fontsize=14)
    
    if trajectory_points == 'endpoints':
        plt.title(f"UMAP Visualization of Initial and Final States (n={len(all_labels)})", fontsize=16)
    elif trajectory_points == 'initial':
        plt.title(f"UMAP Visualization of Initial States (n={len(all_labels)})", fontsize=16)
    else:
        plt.title(f"UMAP Visualization of Final States (n={len(all_labels)})", fontsize=16)
    
    # Add a text box with summary statistics
    n_cad = sum(all_labels == 1)
    n_non_cad = sum(all_labels == 0)
    plt.figtext(0.02, 0.02, 
               f"Total patients: {len(all_labels)}\nCAD patients: {n_cad} ({n_cad/len(all_labels):.1%})\n" +
               f"Non-CAD patients: {n_non_cad} ({n_non_cad/len(all_labels):.1%})",
               fontsize=10, bbox=dict(facecolor='white', alpha=0.7))
    
    plt.legend()
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Figure saved to {save_path}")
        
    if display_in_notebook:
        from IPython.display import display
        display(plt.gcf())
    
    # Create an additional plot showing density
    try:
        import seaborn as sns
        
        plt.figure(figsize=(14, 10))
        
        # Plot density contours for each class
        for label_val in np.unique(umap_labels):
            # For endpoints, only use final states for density
            if trajectory_points == 'endpoints':
                mask = (umap_labels == label_val) & (point_types == 'final')
            else:
                mask = umap_labels == label_val
                
            sns.kdeplot(
                x=embedding[mask, 0], 
                y=embedding[mask, 1],
                fill=True, 
                alpha=0.3, 
                color=colors[label_val],
                levels=5,
                label=color_names[label_val]
            )
            
            # Overlay scatter plot
            plt.scatter(
                embedding[mask, 0], 
                embedding[mask, 1],
                color=colors[label_val], 
                alpha=0.6, 
                s=30
            )
        
        plt.xlabel('UMAP Dimension 1', fontsize=14)
        plt.ylabel('UMAP Dimension 2', fontsize=14)
        plt.title(f"Density Map in UMAP Space (n={len(all_labels)})", fontsize=16)
        plt.legend()
        
        if save_path:
            density_path = save_path[:-4] + "_density.png" if isinstance(save_path, str) else "umap_density.png"
            plt.savefig(density_path, dpi=300, bbox_inches='tight')
            print(f"Density map saved to {density_path}")
            
        if display_in_notebook:
            display(plt.gcf())
            
    except Exception as e:
        print(f"Error creating density plot: {e}")
        print("Make sure seaborn is installed: pip install seaborn")
    
    plt.close('all')
    
    return embedding, umap_labels
```

```python
def visualize_parameter_grid(model, dataloader, args, n_traj=1, display_in_notebook=True, save_path=None):
    """
    Create a grid of visualizations with different UMAP parameters.
    """
    import numpy as np
    import torch
    import matplotlib.pyplot as plt
    from sklearn.manifold import TSNE
    import umap
    
    # Get latent states
    all_final_states = []
    all_labels = []
    
    with torch.no_grad():
        for batch_idx, data_batch in enumerate(dataloader):
            if batch_idx >= 5:  # Limit to 5 batches
                break
                
            # Get latent states
            pred_y, info = model.get_recons(
                tps_to_pred=data_batch["tps_to_pred"],
                obs_data=data_batch["obs_data"],
                obs_tps=data_batch["obs_tps"],
                mask=data_batch["obs_mask"],
                n_traj=n_traj,
                z_last=args.z_last if hasattr(args, 'z_last') else False
            )
            
            # Extract final states
            latent_traj = info["latent_traj"]
            final_states = latent_traj[0, :, -1, :].cpu().numpy()
            
            # Get labels
            labels = data_batch["labels"].cpu().numpy()
            if labels.ndim > 1:
                labels = labels.squeeze()
            
            all_final_states.append(final_states)
            all_labels.append(labels)
    
    # Combine data
    all_final_states = np.concatenate(all_final_states, axis=0)
    all_labels = np.concatenate(all_labels, axis=0)
    
    # Create a grid of visualizations
    fig, axs = plt.subplots(2, 3, figsize=(18, 12))
    axs = axs.flatten()
    
    # t-SNE visualization
    print("Computing t-SNE...")
    tsne = TSNE(n_components=2, random_state=42)
    tsne_result = tsne.fit_transform(all_final_states)
    
    axs[0].scatter(tsne_result[:, 0], tsne_result[:, 1], c=all_labels, cmap='coolwarm', s=30, alpha=0.7)
    axs[0].set_title('t-SNE Visualization')
    
    # UMAP with different parameters
    umap_params = [
        {'n_neighbors': 5, 'min_dist': 0.01, 'title': 'UMAP (n_neighbors=5, min_dist=0.01)'},
        {'n_neighbors': 15, 'min_dist': 0.1, 'title': 'UMAP (n_neighbors=15, min_dist=0.1)'},
        {'n_neighbors': 30, 'min_dist': 0.3, 'title': 'UMAP (n_neighbors=30, min_dist=0.3)'},
        {'n_neighbors': 50, 'min_dist': 0.5, 'title': 'UMAP (n_neighbors=50, min_dist=0.5)'},
        {'n_neighbors': 100, 'min_dist': 0.1, 'title': 'UMAP (n_neighbors=100, min_dist=0.1)'}
    ]
    
    for i, params in enumerate(umap_params):
        print(f"Computing UMAP with {params['title']}...")
        reducer = umap.UMAP(n_components=2, random_state=42, 
                          n_neighbors=params['n_neighbors'], 
                          min_dist=params['min_dist'])
        embedding = reducer.fit_transform(all_final_states)
        
        axs[i+1].scatter(embedding[:, 0], embedding[:, 1], c=all_labels, cmap='coolwarm', s=30, alpha=0.7)
        axs[i+1].set_title(params['title'])
    
    # Adjust layout
    plt.tight_layout()
    
    if display_in_notebook:
        plt.show()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
```

```python
def visualize_latent_space_umap_3d(model, dataloader, args, n_traj=1, 
                                  trajectory_points='endpoints', n_neighbors=15, min_dist=0.1,
                                  display_in_notebook=True, save_path=None):
    """
    Create an interactive 3D UMAP visualization of latent states.
    
    Args:
        model: The trained LatentODE model
        dataloader: DataLoader with patient data
        args: Command line arguments
        n_traj: Number of trajectories to sample
        trajectory_points: Which points to use from trajectories: 'endpoints', 'initial', or 'final'
        n_neighbors: UMAP parameter - controls how UMAP balances local versus global structure
        min_dist: UMAP parameter - controls how tightly UMAP packs points together
        display_in_notebook: Whether to display in notebook
        save_path: Path to save figure
    """
    import numpy as np
    import torch
    from tqdm import tqdm
    
    # Import required libraries
    try:
        import umap
        import plotly.graph_objects as go
        import plotly.express as px
        from plotly.subplots import make_subplots
    except ImportError:
        print("Required libraries not installed. Please install with:")
        print("pip install umap-learn plotly")
        return
    
    model.eval()
    
    # Collect embeddings and labels
    all_initial_states = []
    all_final_states = []
    all_labels = []
    all_probabilities = []  # Store model's prediction probabilities
    all_patient_ids = []
    
    print("Collecting latent representations from all patients...")
    batch_count = 0
    max_batches = 10  # Limit the number of batches for memory reasons
    
    with torch.no_grad():
        for batch_idx, data_batch in enumerate(tqdm(dataloader)):
            if batch_count >= max_batches:
                break
                
            # Get predictions and latent representations
            pred_y, info = model.get_recons(
                tps_to_pred=data_batch["tps_to_pred"],
                obs_data=data_batch["obs_data"],
                obs_tps=data_batch["obs_tps"],
                mask=data_batch["obs_mask"],
                n_traj=n_traj,
                z_last=args.z_last if hasattr(args, 'z_last') else False
            )
            
            # Extract states
            fp_mu, fp_std, fp_enc = info["fp"]
            latent_traj = info["latent_traj"]
            
            # Extract initial and final states
            initial_states = fp_enc[0].cpu().numpy()
            final_states = latent_traj[0, :, -1, :].cpu().numpy()
            
            # Get labels
            if "labels" in data_batch and data_batch["labels"] is not None:
                labels = data_batch["labels"].cpu().numpy()
                if labels.ndim > 1:
                    labels = labels.squeeze()
            else:
                labels = np.zeros(fp_enc.shape[1])
            
            # Get prediction probabilities
            if "label_pred" in info and model.classif:
                pred_probs = torch.sigmoid(info["label_pred"]).cpu().numpy().squeeze()
                if pred_probs.ndim > 1:
                    pred_probs = pred_probs[0]  # Take first trajectory
            else:
                pred_probs = np.zeros(len(labels))
            
            # Extract patient IDs if available
            if "patient_ids" in data_batch:
                patient_ids = data_batch["patient_ids"]
            else:
                # Create dummy IDs
                patient_ids = [f"P{batch_idx}_{i}" for i in range(fp_enc.shape[1])]
            
            # Store data
            all_initial_states.append(initial_states)
            all_final_states.append(final_states)
            all_labels.append(labels)
            all_probabilities.append(pred_probs)
            all_patient_ids.extend(patient_ids)
            
            batch_count += 1
    
    # Combine data from all batches
    all_initial_states = np.concatenate(all_initial_states, axis=0)
    all_final_states = np.concatenate(all_final_states, axis=0)
    all_labels = np.concatenate(all_labels, axis=0)
    all_probabilities = np.concatenate(all_probabilities, axis=0)
    
    # Prepare data for UMAP based on which points we want to use
    if trajectory_points == 'endpoints':
        # Use both initial and final states
        umap_data = np.vstack([all_initial_states, all_final_states])
        umap_labels = np.concatenate([all_labels, all_labels])
        umap_probs = np.concatenate([all_probabilities, all_probabilities])
        point_types = np.array(['initial'] * len(all_labels) + ['final'] * len(all_labels))
        patient_indices = np.concatenate([np.arange(len(all_labels)), np.arange(len(all_labels))])
    elif trajectory_points == 'initial':
        # Use only initial states
        umap_data = all_initial_states
        umap_labels = all_labels
        umap_probs = all_probabilities
        point_types = np.array(['initial'] * len(all_labels))
        patient_indices = np.arange(len(all_labels))
    elif trajectory_points == 'final':
        # Use only final states
        umap_data = all_final_states
        umap_labels = all_labels
        umap_probs = all_probabilities
        point_types = np.array(['final'] * len(all_labels))
        patient_indices = np.arange(len(all_labels))
    else:
        raise ValueError("trajectory_points must be 'endpoints', 'initial', or 'final'")
    
    print(f"Applying UMAP to {umap_data.shape[0]} data points with dimension {umap_data.shape[1]}...")
    
    # Apply UMAP for 3D dimensionality reduction
    reducer = umap.UMAP(n_neighbors=n_neighbors, min_dist=min_dist, n_components=3, random_state=42)
    embedding = reducer.fit_transform(umap_data)
    
    # Create an interactive 3D plot with Plotly
    fig = make_subplots(rows=1, cols=1, specs=[[{'type': 'scatter3d'}]])
    
    # Color scheme
    colors = {0: 'blue', 1: 'red'}  # 0: No CAD, 1: CAD
    color_names = {0: 'Non-CAD', 1: 'CAD'}
    
    # Create the main scatter plot
    for label_val in np.unique(umap_labels):
        for point_type in np.unique(point_types):
            mask = (umap_labels == label_val) & (point_types == point_type)
            
            # Hover text with prediction probabilities
            hover_text = [f"Patient {all_patient_ids[patient_indices[i]] if i < len(all_patient_ids) else 'Unknown'}<br>" + 
                         f"True label: {label_val}<br>" +
                         f"Predicted prob: {umap_probs[i]:.3f}<br>" +
                         f"Point type: {point_type}" 
                         for i in np.where(mask)[0]]
            
            # Marker properties
            marker = dict(
                size=8,
                color=colors[label_val],
                opacity=0.7,
                symbol='circle' if point_type == 'initial' else 'square'
            )
            
            fig.add_trace(
                go.Scatter3d(
                    x=embedding[mask, 0],
                    y=embedding[mask, 1],
                    z=embedding[mask, 2],
                    mode='markers',
                    marker=marker,
                    text=hover_text,
                    hoverinfo='text',
                    name=f"{color_names[label_val]} ({point_type})"
                )
            )
    
    # If using endpoints, add lines connecting initial and final states
    if trajectory_points == 'endpoints':
        n_patients = len(all_labels)
        for i in range(n_patients):
            # Get indices for this patient's initial and final states
            initial_idx = i
            final_idx = i + n_patients
            
            # Only draw lines for a subset of patients to avoid visual clutter
            if i % 5 == 0:  # Draw every 5th patient
                fig.add_trace(
                    go.Scatter3d(
                        x=[embedding[initial_idx, 0], embedding[final_idx, 0]],
                        y=[embedding[initial_idx, 1], embedding[final_idx, 1]],
                        z=[embedding[initial_idx, 2], embedding[final_idx, 2]],
                        mode='lines',
                        line=dict(color=colors[all_labels[i]], width=2, dash='dash'),
                        showlegend=False
                    )
                )
    
    # Update layout
    fig.update_layout(
        title=f"3D UMAP Visualization of Latent States (n={len(all_labels)})",
        scene=dict(
            xaxis_title="UMAP Dimension 1",
            yaxis_title="UMAP Dimension 2",
            zaxis_title="UMAP Dimension 3",
        ),
        legend_title="Patient Groups",
        width=900,
        height=700,
        margin=dict(l=0, r=0, b=0, t=40)
    )
    
    # Add text annotation with statistics
    n_cad = sum(all_labels == 1)
    n_non_cad = sum(all_labels == 0)
    stats_text = (
        f"Total patients: {len(all_labels)}<br>" +
        f"CAD patients: {n_cad} ({n_cad/len(all_labels):.1%})<br>" +
        f"Non-CAD patients: {n_non_cad} ({n_non_cad/len(all_labels):.1%})"
    )
    
    fig.add_annotation(
        x=0.02, y=0.02,
        xref="paper", yref="paper",
        text=stats_text,
        showarrow=False,
        bgcolor="white",
        opacity=0.8,
        bordercolor="black",
        borderwidth=1,
        borderpad=4
    )
    
    # Display or save figure
    if display_in_notebook:
        fig.show()
    
    if save_path:
        fig.write_html(f"{save_path}.html")
        print(f"Interactive 3D plot saved to {save_path}.html")
    
    return embedding, umap_labels
```

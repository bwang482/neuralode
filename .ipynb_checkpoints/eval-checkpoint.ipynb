```python
def parse_model_results(results_text):
    # Initialize a list to store the data
    data = []
    
    # Split the text by model blocks
    model_blocks = results_text.split("Model: ")
    model_blocks = [block.strip() for block in model_blocks if block.strip()]
    
    for model_block in model_blocks:
        lines = model_block.split('\n')
        model_name = lines[0]
        
        # Initialize a dictionary for this model's data
        model_data = {"Model": model_name}
        
        # Extract metrics for this model
        for line in lines[1:]:
            if ":" in line:
                metric, value = line.split(":", 1)
                metric = metric.strip()
                value = float(value.strip())
                
                # Format column name based on metric type
                if "@Specificity-" in metric:
                    base_metric, specificity = metric.split("@Specificity-")
                    base_metric = base_metric.strip()
                    specificity = specificity.strip()
                    column_name = f"{base_metric}_{specificity}"
                elif "risk-" in metric:
                    specificity = metric.split("-")[1].strip()
                    column_name = f"Relative_risk_{specificity}"
                else:
                    column_name = metric.replace(" ", "_")
                
                model_data[column_name] = value
        
        data.append(model_data)
    
    return pd.DataFrame(data)

# Your results text (you would paste your actual results here)
results_text = """
Model: LR
NPV @Specificity-0.80 : 0.6344366711913961
NPV @Specificity-0.90 : 0.5854097097354225
NPV @Specificity-0.95 : 0.5551026686149351
PPV @Specificity-0.80 : 0.7291332977398114
PPV @Specificity-0.90 : 0.7836224054591981
PPV @Specificity-0.95 : 0.8267213862289101
Sensitivity @Specificity-0.80 : 0.5392208475914715
Sensitivity @Specificity-0.90 : 0.36272703342984997
Sensitivity @Specificity-0.95 : 0.23861542511187153
Relative risk-0.80 : 1.4582665954796228
Relative risk-0.90 : 1.5672448109183963
Relative risk-0.95 : 1.6534427724578202
Overall AUROC : 0.7519122337994905

Model: NB
NPV @Specificity-0.80 : 0.6006522383634747
NPV @Specificity-0.90 : 0.5578320137131662
NPV @Specificity-0.95 : 0.5328262314452403
PPV @Specificity-0.80 : 0.7006105968091393
PPV @Specificity-0.90 : 0.7405772495755518
PPV @Specificity-0.95 : 0.7685800604229607
Sensitivity @Specificity-0.80 : 0.46814951302974467
Sensitivity @Specificity-0.90 : 0.2870492234798631
Sensitivity @Specificity-0.95 : 0.16741247696762307
Relative risk-0.80 : 1.4012211936182786
Relative risk-0.90 : 1.4811544991511036
Relative risk-0.95 : 1.5371601208459214
Overall AUROC : 0.7137280986718704

Model: RF
NPV @Specificity-0.80 : 0.6575386946639247
NPV @Specificity-0.90 : 0.6049186128803963
NPV @Specificity-0.95 : 0.5694674556213017
PPV @Specificity-0.80 : 0.7443343965083096
PPV @Specificity-0.90 : 0.8047276464542652
PPV @Specificity-0.95 : 0.8492661642205473
Sensitivity @Specificity-0.80 : 0.5835746249012899
Sensitivity @Specificity-0.90 : 0.4122137404580153
Sensitivity @Specificity-0.95 : 0.2817846801789945
Relative risk-0.80 : 1.4886687930166191
Relative risk-0.90 : 1.6094552929085304
Relative risk-0.95 : 1.6985323284410947
Overall AUROC : 0.7736316795105679

Model: LGBM
NPV @Specificity-0.80 : 0.620996623350046
NPV @Specificity-0.90 : 0.5784529451591063
NPV @Specificity-0.95 : 0.5490225907051038
PPV @Specificity-0.80 : 0.7180527383367139
PPV @Specificity-0.90 : 0.7742603550295858
PPV @Specificity-0.95 : 0.8145436798438263
Sensitivity @Specificity-0.80 : 0.512503290339563
Sensitivity @Specificity-0.90 : 0.3444327454593314
Sensitivity @Specificity-0.95 : 0.2196630692287444
Relative risk-0.80 : 1.4361054766734278
Relative risk-0.90 : 1.5485207100591716
Relative risk-0.95 : 1.6290873596876525
Overall AUROC : 0.744935484054247

Model: Latent-ODE
NPV @Specificity-0.80 : 0.6424079065588498
NPV @Specificity-0.90 : 0.5963978893201343
NPV @Specificity-0.95 : 0.5577837023298945
PPV @Specificity-0.80 : 0.7348966960160405
PPV @Specificity-0.90 : 0.7962739579144886
PPV @Specificity-0.95 : 0.8315229439148748
Sensitivity @Specificity-0.80 : 0.5547512503290339
Sensitivity @Specificity-0.90 : 0.3909581468807581
Sensitivity @Specificity-0.95 : 0.2468412740194788
Relative risk-0.80 : 1.469793392032081
Relative risk-0.90 : 1.5925479158289773
Relative risk-0.95 : 1.6630458878297496
Overall AUROC : 0.7581591153821183
"""

# Create the DataFrame
df = parse_model_results(results_text)
```

```python
# Define plot metrics
metrics = {
    (0, 0): "PPV_0.95",
    (0, 1): "Sensitivity_0.95",
    (1, 0): "Overall_AUROC",
    (1, 1): "Relative_risk_0.95"
}

# Define pretty labels for the plots
labels = {
    "PPV_0.95": "PPV",
    "Sensitivity_0.95": "Sensitivity",
    "Overall_AUROC": "AUROC",
    "Relative_risk_0.95": "Relative Risk"
}

y_labels = {
    "PPV_0.95": "PPV",
    "Sensitivity_0.95": "Sensitivity",
    "Overall_AUROC": "AUROC",
    "Relative_risk_0.95": "Relative Risk"
}

# Set Seaborn style
sns.set(style="darkgrid")
colors = sns.color_palette("tab10", n_colors=len(df))

# Set up the plot
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
plt.subplots_adjust(hspace=0.3, wspace=0.3)

# Create bar plots using Seaborn
for (i, j), metric in metrics.items():
    ax = axes[i, j]
    sns.barplot(
        x='Model', 
        y=metric, 
        data=df, 
        hue='Model',
        palette=colors,
        ax=ax
    )
    
    # Set plot attributes
    # ax.set_title(labels[metric], fontsize=13, fontweight='bold')
    ax.set_xlabel('')  # Remove x-axis label as it's redundant
    ax.set_ylabel(y_labels[metric], fontsize=12)
    # plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

    if i == 0:  # If this is the top row
        ax.set_xticklabels([])  # Remove x-tick labels
    
    # Set y-axis limits appropriate for each metric
    if metric == "Overall_AUROC":
        ax.set_ylim(0.6, 0.8)  # Typical AUROC range
    elif metric == "Sensitivity_0.95":
        ax.set_ylim(0, 0.35)   # Lower values for sensitivity at high specificity
    elif metric == "PPV_0.95":
        ax.set_ylim(0.7, 0.9)  # Higher values for PPV at high specificity
    elif metric == "Relative_risk_0.95":
        ax.set_ylim(1.4, 1.8)   # Range for relative risk
    
    # # Add value labels on top of bars
    for p, value in zip(ax.patches, df[metric].values):
        height = p.get_height()
        ax.text(
            p.get_x() + p.get_width()/2.,
            height + 0.01,
            f'{value:.3f}',
            ha='center', 
            va='bottom', 
            fontsize=9
        )

# Add a single legend at the bottom of the figure
handles = [plt.Rectangle((0,0),1,1, color=colors[i]) for i in range(len(df))]
fig.legend(handles, df['Model'], loc='lower center', ncol=len(df), bbox_to_anchor=(0.5, 0.02), 
           fontsize=11, frameon=True)

fig.suptitle('Model Performance Comparison (CAD)', fontsize=16, fontweight='bold', y=0.98)
fig.tight_layout(rect=[0, 0.05, 1, 0.95])  # Adjust for the legend at the bottom
plt.show()
```
